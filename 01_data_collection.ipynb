```python
import yfinance as yf
from datetime import datetime, timedelta
import numpy as np

def get_nifty_conditions(year):
    """Get Nifty 50 stats for the year of IPO — market sentiment signal"""
    try:
        start = f"{year}-01-01"
        end = f"{year}-12-31"
        nifty = yf.download("^NSEI", start=start, end=end, progress=False)
        
        if nifty.empty:
            return None, None, None
        
        avg_price = nifty["Close"].mean().values[0]
        volatility = nifty["Close"].pct_change().std().values[0] * 100
        yearly_return = ((nifty["Close"].iloc[-1] - nifty["Close"].iloc[0]) / nifty["Close"].iloc[0] * 100).values[0]
        
        return round(float(avg_price), 2), round(float(volatility), 4), round(float(yearly_return), 2)
    except Exception as e:
        print(f"Error for year {year}: {e}")
        return None, None, None

# Get unique years and fetch market data once per year
print("Fetching Nifty 50 market data...")
years = df["year"].unique()
market_data = {}

for year in sorted(years):
    avg, vol, ret = get_nifty_conditions(year)
    market_data[year] = {
        "nifty_avg": avg,
        "nifty_volatility": vol,
        "nifty_yearly_return": ret
    }
    print(f"Year {year} -> Avg: {avg}, Volatility: {vol}%, Return: {ret}%")

print("\nMarket data fetched successfully!")
```

```python
# Add market columns to dataframe
df["nifty_avg"] = df["year"].map(lambda y: market_data[y]["nifty_avg"])
df["nifty_volatility"] = df["year"].map(lambda y: market_data[y]["nifty_volatility"])
df["nifty_yearly_return"] = df["year"].map(lambda y: market_data[y]["nifty_yearly_return"])

# Add derived features
df["listing_gain_pct"] = ((df["listing_price"] - df["issue_price"]) / df["issue_price"] * 100).round(2)
df["issue_size_log"] = np.log1p(df["issue_size_cr"])

# Create target variable — subscription category (what we will predict)
def categorize_subscription(x):
    if x < 1:
        return 0   # Under-subscribed
    elif x < 10:
        return 1   # Low (1x–10x)
    elif x < 50:
        return 2   # Medium (10x–50x)
    else:
        return 3   # High (50x+)

df["subscription_category"] = df["total_sub"].apply(categorize_subscription)

category_labels = {0: "Under-subscribed", 1: "Low (1–10x)", 2: "Medium (10–50x)", 3: "High (50x+)"}
df["subscription_label"] = df["subscription_category"].map(category_labels)

print("Features added successfully!")
print(f"\nNew shape: {df.shape}")
print(f"\nTarget variable distribution:")
print(df["subscription_label"].value_counts())
print(f"\nSample enriched data:")
df[["company", "year", "total_sub", "subscription_label", "nifty_avg", "nifty_yearly_return", "listing_gain_pct"]].head(10)
```
